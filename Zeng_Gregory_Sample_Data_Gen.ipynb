```python
from faker import Faker
import random
from datetime import datetime

fake = Faker()

num_customers = 75

flight_no = []
restaurantID = []
typeID = []
BookedID = []
hotelID = []
paymentID = []
airline_name = []
customerID = []
ticketID = []
classID = []
cuisineID = []
activitiesID = []
navigationID = []
date_time = []

while len(flight_no) < num_customers:
    value = fake.random_number(digits=5)
    if value not in flight_no:
        flight_no.append(value)

while len(restaurantID) < num_customers:
    value = fake.random_int(min=0, max=1000)
    if value not in restaurantID:
        restaurantID.append(value)

while len(typeID) < num_customers:
    value = fake.random_int(min=0, max=1000)
    if value not in typeID:
        typeID.append(value)        
        
while len(BookedID) < num_customers:
    value = fake.random_int(min=0, max=1000)
    if value not in BookedID:
        BookedID.append(value)      

while len(hotelID) < num_customers:
    value = fake.random_int(min=0, max=1000)
    if value not in hotelID:
        hotelID.append(value)
        
while len(paymentID) < num_customers:
    value = fake.random_int(min=0, max=1000)
    if value not in paymentID:
        paymentID.append(value)

while len(airline_name) < num_customers:
    value = fake.company()
    if value not in airline_name:
        airline_name.append(value)

while len(customerID) < num_customers:
    value = fake.random_int(min=0, max=1000)
    if value not in customerID:
        customerID.append(value)

while len(ticketID) < num_customers:
    value = fake.random_int(min=0, max=1000)
    if value not in ticketID:
        ticketID.append(value)

while len(classID) < num_customers:
    value = fake.random_int(min=0, max=1000)
    if value not in classID:
        classID.append(value)

while len(cuisineID) < num_customers:
    value = fake.random_int(min=0, max=1000)
    if value not in cuisineID:
        cuisineID.append(value)

while len(activitiesID) < num_customers:
    value = fake.random_int(min=0, max=1000)
    if value not in activitiesID:
        activitiesID.append(value)

while len(navigationID) < num_customers:
    value = fake.random_int(min=0, max=1000)
    if value not in navigationID:
        navigationID.append(value)

while len(date_time) < num_customers:
    value = datetime.strftime(fake.date_time(), '%Y-%m-%d %H:%M:%S')
    if value not in date_time:
        date_time.append(value)
```
